# `Tensors`

```python
import numpy as np
import torch
```

## **NumPy** `Array`

```python
array = np.zeros((3,100))
print(array.dtype)
```

## **PyTorch** `tensor`

```python
tensor1 = torch.zeros((3,100))
print(tensor1.dtype)
```

## `tensor` **operations**

```python
tensor2 = torch.tensor(array)
print(tensor2.dtype)
```

```python
tensor3 = tensor2.reshape(2, -1)
tensor3.shape
```

```python
tensor3[0][0] = 1
tensor2[0]
```

## `tensor` **math**

```python
torch.sub(tensor1 + tensor2, tensor1 - tensor2)
```

```python
torch.tensor(np.array([i for i in range(150)])) ** torch.arange(150)
```
